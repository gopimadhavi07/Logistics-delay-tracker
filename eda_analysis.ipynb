# Logistics Delay Tracker EDA
This notebook performs exploratory data analysis on `logistics_features.csv` and generates visualizations saved to the `charts/` folder.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

# Load dataset
data_path = os.path.join('data', 'logistics_features.csv')
df = pd.read_csv(data_path, parse_dates=['promised_date', 'delivery_date'])
# Ensure charts directory exists
charts_dir = 'charts'
os.makedirs(charts_dir, exist_ok=True)

# Set plot style
sns.set_style('whitegrid')
```

```python
# 1. Order Volume Distribution by Month
order_month = df['month'].value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.barplot(x=order_month.index, y=order_month.values, palette='viridis')
plt.title('Order Volume by Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'order_volume_by_month.png'))
plt.show()
```

```python
# 2. Delay Duration Distribution (Days)
plt.figure(figsize=(8,5))
sns.histplot(df['delay_duration'], bins=range(int(df['delay_duration'].min()), int(df['delay_duration'].max())+2), kde=False, color='steelblue')
plt.title('Delay Duration Distribution')
plt.xlabel('Delay Duration (days)')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'delay_duration_hist.png'))
plt.show()
```

```python
# 3. Delay Percentage by Courier
courier_delay = df.groupby('courier')['is_delayed'].mean().reset_index().rename(columns={'is_delayed':'delay_pct'})
plt.figure(figsize=(8,5))
sns.barplot(data=courier_delay, x='courier', y='delay_pct', palette='coolwarm')
plt.title('Delay Percentage by Courier')
plt.ylabel('Delay %')
plt.ylim(0,1)
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'delay_pct_by_courier.png'))
plt.show()
```

```python
# 4. Delay Percentage by Vendor
vendor_delay = df.groupby('vendor')['is_delayed'].mean().reset_index().rename(columns={'is_delayed':'delay_pct'})
plt.figure(figsize=(8,5))
sns.barplot(data=vendor_delay, x='vendor', y='delay_pct', palette='magma')
plt.title('Delay Percentage by Vendor')
plt.ylabel('Delay %')
plt.ylim(0,1)
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'delay_pct_by_vendor.png'))
plt.show()
```

```python
# 5. Monthly Delay Trend (Average Delay)
monthly_delay = df.groupby('month')['delay_duration'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.lineplot(data=monthly_delay, x='month', y='delay_duration', marker='o')
plt.title('Average Delay Duration by Month')
plt.xlabel('Month')
plt.ylabel('Average Delay (days)')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'avg_delay_by_month.png'))
plt.show()
```

```python
# 6. Revenue Impact (Total Revenue for Delayed Orders)
revenue_impact = df.loc[df['is_delayed'], 'revenue_impacted'].sum()
print(f'Total Revenue Impact from Delayed Orders: ${revenue_impact:,.2f}')
# Save as a simple text plot
plt.figure(figsize=(6,1))
plt.text(0.5, 0.5, f'Total Revenue Impact: ${revenue_impact:,.2f}', horizontalalignment='center', fontsize=12)
plt.axis('off')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'revenue_impact.png'))
plt.show()
```

```python
# 7. Delay Category Distribution
cat_counts = df['delay_category'].value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.barplot(x=cat_counts.index, y=cat_counts.values, palette='pastel')
plt.title('Delay Category Distribution')
plt.xlabel('Delay Category')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'delay_category_distribution.png'))
plt.show()
```
